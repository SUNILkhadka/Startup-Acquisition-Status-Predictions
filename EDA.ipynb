## Ploting & Exploring features 

```python
# Checking weather there is relationship between some important features or not 
num_column = ['funding_rounds', 'funding_total_usd', 'milestones', 'relationships','isClose', 'activeDays']
sns.set_palette('dark')
sns.pairplot(company[num_column])
plt.show()
```

```python
# Plot showing distribution of data 
for i in range(len(num_column)):
    plt.figure(figsize=(15,6))
    sns.kdeplot(company[num_column[i]],color='green',shade=True,thresh=True)
    plt.show()
```

```python
# Boxplot of funding total usd
plt.figure(figsize=(10,6))
sns.boxplot(x='funding_total_usd',data=company,palette='dark')
plt.title('Box plot')
plt.show()
```

```python
# Plotting pie charts of isClose column
plt.figure(figsize=(8,6))
plt.pie(company.isClose.value_counts(),autopct='%.2f',labels=company.isClose.value_counts().index)
plt.title('Pie Charts showing Company status',color='blue')
plt.show()
```

```python
# Boxplot of activeDays 
plt.figure(figsize=(15,8))
sns.boxplot(x='activeDays',data=company,linewidth=3,fliersize=3,saturation=1)
plt.show()
```

isActive coluns data are positively skewed

#### Checking relationship between columns

```python
# Calculating coefficient of correlations
corr_coef = company.corr()
plt.figure(figsize=(30,30))
sns.heatmap(corr_coef,fmt='.2f',annot=True)
plt.show()
```
